```python
NY_News_Details  = []
for bigbox in bigboxes:
    try:
        news_link = "https://www.nytimes.com/" + bigbox.find("div", {"class": "css-1l4spti"}).a['href']
    except:
        news_link = "No News Link"
        
    try:
        news_headlines = bigbox.find("h2", {"class": "css-1j9dxys e1xfvim30"}).text
    except:
        news_headlines = "No News Headlines"
        
    newsRes = requests.get(news_link)
    newsRes.encoding = 'utf-8'
    news_content = bs(newsRes.text, "html.parser")
    
    try: 
        date_published = news_content.find("time", {"class": "css-1sbuyqj e16638kd3"}).text
        
    except:
        date_published = "No Date Published"
        
    try: 
        news_body = news_content.findAll("div", {"class": "css-1fanzo5 StoryBodyCompanionColumn"})
    except:
        news_body = "No News Body"
    
    news_description=""
    
    try:
        for news_desc in news_body:
            paragraphs = news_desc.findAll('p')
            for paragraph in paragraphs:
                paragraph = paragraph.get_text()
                news_description = news_description + paragraph
    except:
        news_description = "No News Description"
    
    mydict = {"News Link": news_link, "News Headlines": news_headlines, "Date Published" : date_published, "News Description":news_description, "Domain": "Miscellaneous", "Company": "Miscellaneous"}  
    NY_News_Details.append(mydict)
```

```python
import nltk as nltk
import re
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
## Remove Punctuation
def remove_punctuation(text): #Remove punctuations
    text_nopunct = ''
    #text_nopunct = re.sub('['+string.punctuation+']', ' ', text)
    text_nopunct = re.sub('[^a-zA-Z]', ' ', text)
    return text_nopunct

NY_News_DataFrame['News_Description_Cleaned'] = NY_News_DataFrame['News_Description'].apply(lambda News_Headline: remove_punctuation(News_Description))
```

```python
NY_News_DataFrame
```

```python
## Tokenize sentences
tokens = [word_tokenize(sen) for sen in NY_News_DataFrame.News_Headlines_Cleaned]
```

```python
## Convert to lower case
def lowercase_token(tokens): 
    return [w.lower() for w in tokens]    
    
lower_tokens = [lowercase_token(token) for token in tokens]
```

```python
## Remove stopwords
stoplist = stopwords.words('english')
def removestopwords(tokens): 
    return [word for word in tokens if word not in stoplist]
filtered_words = [removestopwords(sen) for sen in lower_tokens]
```

```python
## Lematization of words
def lemmatize_words(filtered_words):
    lemmatizer=WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in filtered_words]
lemmatize_words = [lemmatize_words(sen) for sen in filtered_words]

def lemmatize_verbs(lemmatize_words):
    lemmatizer=WordNetLemmatizer()
    return [lemmatizer.lemmatize(word, pos='v') for word in lemmatize_words]
lemmatize_verbs = [lemmatize_verbs(sen) for sen in lemmatize_words]
```

```python
NY_News_DataFrame['News_Headlines_Cleaned'] = [' '.join(sen) for sen in lemmatize_verbs]
NY_News_DataFrame['tokens'] = lemmatize_verbs
NY_News_DataFrame
```

```python
NY_News_DataFrame.to_excel(excel_writer='NY_NEWS.xlsx',sheet_name="NYTIMES_NEWS")
```

```python
from datetime import datetime
from dateutil.parser import parse
import pandas as pd
```

```python
AIB_urls = []
for page_no in range(3):
    AIB_urls.append("https://aibusiness.com/archives.asp?section_id=782&piddl_archivepage={}".format(page_no))
    
AIB_urls = AIB_urls[1:]
```

```python
AIB_NEWS_DataFrame = pd.DataFrame()
for AIB_url in AIB_urls:
    req = urllib2.Request(AIB_url, headers=hdr)
    uClient = uReq(req) # requesting the webpage from the internet
    news_page = uClient.read() # reading the webpage
    uClient.close
    AIB_html = bs(news_page, "html.parser")
    news_links = []
    news_headlines = []
    dates_published = []
    news_description = []
    for link in AIB_html.findAll("h5"):
        try:
            news_link = "https://aibusiness.com/" + link.a['href']
            news_links.append(news_link)
        except:
            news_link = "No News Link"
        try:
            news_headline = link.a.text
            news_headlines.append(news_headline)
        except:
            news_headline = "No News HeadLines"
    for time in AIB_html.findAll("time"):
        try:
            date_published = time.text
            date_published = parse(date_published).strftime("%d %b %Y ")
            dates_published.append(date_published)
        except:
            date_published = "No date published"
    for news_link in news_links:
        newsRes = requests.get(news_link)
        newsRes.encoding = 'utf-8'
        news_content = bs(newsRes.text, "html.parser")
        news_body = ""
        try: 
            for paragraph in news_content.findAll('p')[2:]:
                paragraph = paragraph.get_text()
                news_body = news_body + paragraph
                
        except:
            news_description = "No News Description"
            
        news_description.append(news_body)
    mydict = {"News Link": news_links, "News Headlines": news_headlines, "Date Published" : dates_published[:-1], "News Description":news_description, "Domain": "Miscellaneous", "Company": "Miscellaneous"}    
    AIB_NEWS_DataFrame = pd.concat([AIB_NEWS_DataFrame,pd.DataFrame(mydict)])
```

```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import pandas as pd
import nltk as nltk
import re
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
def extract_AIDAILY_html(url):
    try:
        uClient = uReq(url) # requesting the webpage from the internet
        news_page = uClient.read() # reading the webpage
        uClient.close
        AIDAILY_html = bs(news_page, "html.parser")
    except:
        AIDAILY_html = "Webpage Not Responding. Try Including Header"
        
    
    return AIDAILY_html
```

```python
def get_Domain_Links(html):
    try:
        domains_html = html.findAll("a",{"class":"Mobile-overlay-folder-item"})
    except:
        domains_html = " No Domain html Avialable"
    try:
        News_Domains = []
        for domain in domains_html:
            domain_name = domain.text.strip()
            News_Domains.append(domain_name)
    
        News_Domains = News_Domains[1:]
    except:
        News_Domains = "No Domains Retrived"
    try:
        News_Domain_Links = []
        for news_domain in News_Domains:
            News_Domain_Link = "https://aidaily.co.uk/"+news_domain.lower()
            News_Domain_Links.append(News_Domain_Link)
    except:
        News_Domain_Links = "Domain-wise News Link Not available"
    return News_Domain_Links
```

```python
def get_NewsHeadlines(summary_content):
    try:
        news_headline = summary_content.find("div",{"class":"summary-title"}).a.text
    except:
        news_headline = "No News Headlines"
    return news_headline
```

```python
def get_NewsLinks(summary_content):
    try:
        news_link = "https://aidaily.co.uk"+summary_content.find("div",{"class":"summary-title"}).a['href']
    except:
        news_link = "No News Link"
    return news_link
```

```python
def get_PublishedDate(summary_content):
    try:
        published_date = summary_content.find("div",{"class":"summary-metadata summary-metadata--primary"}).time.text
    except:
        published_date = "No Published Date"
    return published_date
```

```python
def get_NewsDescriptions(news_content):
    news_body = ""
    try:
        for paragraph in news_content.findAll('p'):
                paragraph = paragraph.get_text()
                news_body = news_body + paragraph
    except:
        news_body = "No News Description"
    return news_body
```

```python
## Remove Punctuation
def remove_punctuation(text):
    try:
        text_nopunct = ''
        #text_nopunct = re.sub('['+string.punctuation+']', ' ', text)
        text_nopunct = re.sub('[^A-Za-z0-9]+', ' ', text)
    except:
        text_nopunct = "Text punctuation removal not working"
    return text_nopunct
```

```python
## Convert to lower case
def lowercase_token(tokens): 
    return [w.lower() for w in tokens]
```

```python
## Remove stopwords
def removestopwords(tokens, stoplist): 
    return [word for word in tokens if word not in stoplist]
```

```python
## Lematization of words
def process_lemmatize_words(filtered_words):
    lemmatizer=WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in filtered_words]
```

```python
## Lematization of words verbs
def process_lemmatize_verbs(lemmatize_words):
    lemmatizer=WordNetLemmatizer()
    return [lemmatizer.lemmatize(word, pos='v') for word in lemmatize_words]
```

```python
def save_News_to_excel(AIDAILY_NEWS_DataFrame,FileName):
    try:
        AIDAILY_NEWS_DataFrame.to_excel(excel_writer= FileName+".xlsx",sheet_name=FileName)
        print("Files Saved Successfully")
    except:
        print("Error occured in saving the file.")
```

```python
def AIDAILY_News(AIDAILY_URL, FileName):
    AIDAILY_NEWS_DataFrame = pd.DataFrame()
    AIDAILY_html = extract_AIDAILY_html(AIDAILY_URL)
    Domain_Links = get_Domain_Links(AIDAILY_html)
    for Domain_Link in Domain_Links:
        try:
            newsRes = requests.get(Domain_Link)
            newsRes.encoding = 'utf-8'
            news_article = bs(newsRes.text, "html.parser")
        except:
            print("Domain-wise news-article page not avialable")
        try:
            summary_contents = news_article.findAll("div",{"class":"summary-content sqs-gallery-meta-container"})
        except:
            summary_contents = "No Domain-wise sumarry content available"
        News_Headlines = []
        News_Links = []
        Dates_Published = []
        Domains = []
        News_Descriptions = []
        for summary_content in summary_contents:
            #Get List Of News HeadLines
            news_headline = get_NewsHeadlines(summary_content)
            News_Headlines.append(news_headline)
            
            #Get List Of News Links
            news_link = get_NewsLinks(summary_content)
            News_Links.append(news_link)
            
            #Get Published Date
            published_date = get_PublishedDate(summary_content)
            Dates_Published.append(published_date)
            
            #Known Domain Included
            domain = Domain_Link[22:]
            Domains.append(domain)
        
        for news_link in News_Links:
            try:
                newsRes = requests.get(news_link)
                newsRes.encoding = 'utf-8'
                news_content = bs(newsRes.text, "html.parser")
            except:
                news_content = "No content for news"
            #Get news Description 
            news_description = get_NewsDescriptions(news_content)
            News_Descriptions.append(news_description)
        
        mydict = {"News Link": News_Links, "News Headlines": News_Headlines, "Date Published" : Dates_Published, "News Description":News_Descriptions, "Domain": Domains, "Company": "Miscellaneous"}    
        AIDAILY_NEWS_DataFrame = pd.concat([AIDAILY_NEWS_DataFrame, pd.DataFrame(mydict, index = None)])
        #print(AIDAILY_NEWS_DataFrame)
    
    AIDAILY_NEWS_DataFrame.sort_values(by = ['News Headlines','Date Published'],  inplace = True)
    AIDAILY_NEWS_DataFrame.drop_duplicates(subset = ['News Headlines','Date Published'], keep = 'first', inplace = True)
    try:
        AIDAILY_NEWS_DataFrame = AIDAILY_NEWS_DataFrame.rename(columns = {'News Description':'News_Description'})
        AIDAILY_NEWS_DataFrame['News_Description_Cleaned'] = AIDAILY_NEWS_DataFrame['News_Description'].apply(lambda News_Description: remove_punctuation(News_Description))
        ## Tokenize sentences
        tokens = [word_tokenize(sen) for sen in AIDAILY_NEWS_DataFrame.News_Description_Cleaned]
        lower_tokens = [lowercase_token(token) for token in tokens]
        stoplist = stopwords.words('english')
        filtered_words = [removestopwords(sen, stoplist) for sen in lower_tokens]
        lemmatize_words = [process_lemmatize_words(sen) for sen in filtered_words]
        lemmatize_verbs = [process_lemmatize_verbs(sen) for sen in lemmatize_words]
        AIDAILY_NEWS_DataFrame['News_Description_Cleaned'] = [' '.join(sen) for sen in lemmatize_verbs]
        AIDAILY_NEWS_DataFrame['tokens'] = lemmatize_verbs
    except:
        print("There is some problem with text pre-processing")
    save_News_to_excel(AIDAILY_NEWS_DataFrame,FileName)
            
```
